# Analyser script

`````python
# Calcule et affiche la proportion de mots de plus de 4 lettres d'une phrase
nombre_min_lettres = 4 # définir un paramètre
sortie_fichier = True # définir un paramètre 
print("Calcul de la proportion de mots avec 4 lettres ou plus.") # afficher une ligne
phrase = input("Écrivez une phrase: ") # demander à l'utilisateur une entrée
liste_mots = phrase.split(" ") # couper la phrase en mots
nombre_mots = len(liste_mots) # on calcule le nombre de mots
if len(phrase) > 0: # on teste la longueur de la phrase
    compteur = 0 # créer une nouvelle variable
    for i in liste_mots: # on fait une boucle sur les mots
        if len(i) >= nombre_min_lettres: # on regarde si le mot dépasse le seuil
            compteur += 1 # on incrémente le compteur
    proportion = round(100 * compteur / nombre_mots, 2) # on calcule la proportion
    informations = {
        "Total": compteur,
        "Proportion": proportion,
        "Phrase": phrase,
        "Seuil": nombre_min_lettres,
    }
    sortie = f"Proportion de mots avec {nombre_min_lettres} lettres ou plus : {proportion}" # mise en forme de la sortie écran
    print(sortie)
    if sortie_fichier:
        with open("resultat.txt", "w") as f:
            f.write(str(informations))
else:
    print("Vous n'avez rien écrit.")
`````

## Modifier le script pour lire un fichier

`````python
nombre_min_lettres = 4
sortie_fichier = True 
print("Calcul de la proportion de mots avec 4 lettres ou plus.")


with open("texte.txt", "r") as monfichier:
    phrase = monfichier.read()
    
# phrase = input("Écrivez une phrase: ") 

liste_mots = phrase.split(" ")
nombre_mots = len(liste_mots)
if len(phrase) > 0:
    compteur = 0 
    for i in liste_mots:
        if len(i) >= nombre_min_lettres: 
            compteur += 1 
    proportion = round(100 * compteur / nombre_mots, 2) 
    informations = {
        "Total": compteur,
        "Proportion": proportion,
        "Phrase": phrase,
        "Seuil": nombre_min_lettres,
    }
    sortie = f"Proportion de mots avec {nombre_min_lettres} lettres ou plus : {proportion}" # mise en forme de la sortie écran
    print(sortie)
    if sortie_fichier:
        with open("resultat.txt", "w") as f:
            f.write(str(informations))
else:
    print("Vous n'avez rien écrit.")
`````
