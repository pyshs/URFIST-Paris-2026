## Utiliser des mod√®les d'HuggingFace : le cas de GliNER

Aller vers les mod√®les :

Un mod√®le plus complexe pour l'identification d'entit√©s

```python
#!pip install gliner
```

```python
from gliner import GLiNER

model = GLiNER.from_pretrained("urchade/gliner_multi")

# Sample text for entity prediction
text = """
L'intelligence articielle est super.
"""

# Labels for entity prediction
labels = ["Domaine"] # for v2.1 use capital case for better performance

# Perform entity prediction
entities = model.predict_entities(text, labels, threshold=0.5)

# Display predicted entities and their labels
for entity in entities:
    print(entity["text"], "=>", entity["label"])
```
