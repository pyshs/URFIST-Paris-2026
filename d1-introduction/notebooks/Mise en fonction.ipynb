# Créer des fonctions

## Logique de fonction

- des entrées
- des opérations faites sur ces entrées
- des sorties

in --> fonction --> out


Renvoyer le premier mot d'une phrase

- entrée : phrase
- opération : il faut récupérer le premier mot
- sortie : renvoyer le mot

```python
phrase = "Ceci est une phrase"
```

```python
phrase.split(" ")[0]
```

```python
def renvoyer_premier_mot(phrase):
    """
    Renvoyer le premier mot
    """
    mot = phrase.split(" ")[0] 
    return mot
```

```python
renvoyer_premier_mot("Ceci est une phrase")
```

## Mettre en fonction notre script

```python
def calculer_nombre_mots_seuil(phrase, nombre_min_lettres = 4):
    """
    Compter le nombre de mots de plus de N lettres
    """
    sortie_fichier = True 
    liste_mots = phrase.split(" ")
    nombre_mots = len(liste_mots)
    if len(phrase) > 0:
        compteur = 0 
        for i in liste_mots: 
            if len(i) >= nombre_min_lettres: 
                compteur += 1
        proportion = round(100 * compteur / nombre_mots, 2) 
        informations = {
            "Total": compteur,
            "Proportion": proportion,
            "Phrase": phrase,
            "Seuil": nombre_min_lettres,
        }
        return informations
    else:
        return None
```

```python
calculer_nombre_mots_seuil("Ceci est une phrase", 4)
```

```python
calculer_nombre_mots_seuil("Ceci est une phrase", 10)
```
